# Data Cleaning

```python
# Calculate the median of the Gross column
gross_median = df['gross'].median()

# Fill the missing values in the Gross column with the median value
df['gross'].fillna(gross_median, inplace=True)
pd.options.display.float_format = '{:.2f}'.format

df
```

```python
# Calculate the median of the budget column
budget_median = df['budget'].median()

# Fill the missing values in the Budget column with the median budget value
df['budget'].fillna(budget_median, inplace=True)
pd.options.display.float_format = '{:.2f}'.format

df
```

```python
# Count the occurrences of each value in the Aspect Ratio column
aspect_ratio_counts = df['aspect_ratio'].value_counts()

# Get the most common value in the Aspect Ratio column
most_common_aspect_ratio = aspect_ratio_counts.idxmax()
df['aspect_ratio'].fillna(most_common_aspect_ratio, inplace=True)

df
```

```python
# Count the occurrences of each value in the content rating column
content_rating_counts = df['content_rating'].value_counts()

# Get the most common value in the content rating column
most_common_content_rating = content_rating_counts.idxmax()
df['content_rating'].fillna(most_common_content_rating, inplace=True)

df
```

# Data Visualization

```python
# Group the movies by content rating and calculate the mean gross revenue for each group
content_rating_means = df.groupby('content_rating')['gross'].mean().sort_values(ascending=False)

# Create a bar plot of mean gross revenue by content rating
plt.bar(content_rating_means.index, content_rating_means.values)
plt.xticks(rotation=90)
plt.xlabel('content_rating')
plt.ylabel('Mean Gross Revenue (in millions of dollars)')
plt.title('Mean Gross Revenue by Genre')
plt.gca().yaxis.set_major_formatter('{:.2f}'.format)
plt.show()
```

Based on the bar chart of mean gross revenue by content rating, it appears that there is a wide variation in mean gross revenue across different content ratings. Specifically, the chart suggests that movies with an "G" rating tend to have the highest mean gross revenue, followed by movies with a "PG" rating, while movies with a "NC-17" rating tend to have the lowest mean gross revenue.

```python
# Group the movies by runtime and calculate the average production budget for each group
avg_budget_by_runtime = df.groupby('duration')['budget'].mean().reset_index()

# Create a line plot of average production budget vs. runtime
plt.plot(avg_budget_by_runtime['duration'], avg_budget_by_runtime['budget'] / 1e6)
plt.xlabel('Runtime (in minutes)')
plt.ylabel('Average Production Budget (in millions of dollars)')
plt.title('Average Movie Production Budget vs. Runtime')
plt.show()
```

Based on the line plot of average production budget vs. runtime, it appears that there is a general trend of increasing production budget with increasing runtime, up to around 150-160 minutes, after which the trend seems to plateau or even decrease slightly.

This suggests that movies with longer runtimes tend to have higher production budgets up to a certain point, after which there may be diminishing returns or other factors that affect the relationship between runtime and production budget

```python
# Create a scatter plot of Budget vs Gross
plt.scatter(df['budget'], df['gross'])
plt.ylabel('gross')
plt.xlabel('budget')
plt.title('Gross vs Budget')

plt.gca().xaxis.set_major_formatter('${:.2f}'.format)
plt.gca().yaxis.set_major_formatter('${:.2f}'.format)
plt.xticks(rotation=45)
plt.xlim(0, 300e6)
plt.ylim(0, 500e6)
plt.show()
```

We can see that there is a general positive correlation between the budget and gross revenue. This means that as the budget of a movie increases, the gross revenue tends to increase as well. However, there are also many outliers and variations in the data, so we can't make any definitive conclusions without further analysis.





```python
# Extract the 'Budget', 'Gross', and 'Director' columns
subset_df = df[['budget', 'gross', 'director_name']]

# Calculate the total gross revenue for each director
director_gross = subset_df.groupby('director_name')['gross'].sum().sort_values(ascending=False)[:10]

# Calculate the total budget for each director
director_budget = subset_df.groupby('director_name')['budget'].sum().sort_values(ascending=False)[:10]

# Create a bar chart of the top 10 directors by gross revenue
plt.bar(director_gross.index, director_gross.values / 1e6)
plt.xticks(rotation=90)
plt.xlabel('Director')
plt.ylabel('Total Gross Revenue (in millions of dollars)')
plt.title('Top 10 Directors by Gross Revenue')
plt.gca().yaxis.set_major_formatter('{:.2f}M'.format)
plt.show()
```

From the chart, we can observe that Steven Spielberg is the top director with the highest total gross revenue. He has directed some of the most popular and successful movies of all time, such as Jurassic Park, E.T. the Extra-Terrestrial, and Jaws. The total gross revenue for Steven Spielberg is over $4 billion.

Other notable directors in the list include Peter Jackson, who directed the Lord of the Rings trilogy and The Hobbit trilogy, and James Cameron, who directed Titanic and Avatar.

This chart shows that there is a clear correlation between a director's success in terms of gross revenue and the popularity of the movies they direct. Successful directors tend to have a strong track record of directing popular movies that attract a large audience and generate significant revenue.





```python
# Compute the total budget and gross revenue for each director
director_budget = df.groupby('director_name')['budget'].sum()
director_gross = df.groupby('director_name')['gross'].sum()

# Get the top 10 directors based on their total gross revenue
top_directors = director_gross.nlargest(10).index

# Create a scatter plot of budget vs. gross revenue for the top 10 directors
plt.scatter(director_gross[top_directors].values / 1e6, director_budget[top_directors].values / 1e6)

for i, director in enumerate(top_directors):
    plt.annotate(director, (director_gross[top_directors].values[i] / 1e6, director_budget[top_directors].values[i] / 1e6))

plt.xlabel('Total Gross Revenue (in millions of dollars)')
plt.ylabel('Total Budget (in millions of dollars)')
plt.title('Gross Revenue vs. Budget for Top 10 Directors')
plt.gca().xaxis.set_major_formatter('{:.2f}'.format)

```

The plot shows that there is a positive correlation between budget and gross revenue for most of the directors. Steven Spielberg and Peter Jackson have the highest total budgets and gross revenue, with Michael Bay coming in a close third. Overall, the plot suggests that having a higher budget is generally associated with higher gross revenue, but there are exceptions to this trend.


```python
# Create the axis object
fig, ax = plt.subplots()

# Group the movies by critic rating and IMDB rating and calculate the average gross revenue for each group
avg_gross_by_rating = df.groupby(['imdb_score', 'critic_reviews'])['gross'].mean().reset_index()

# Set the x-axis formatter
ax.xaxis.set_major_formatter('{:.1f}'.format)

# Sort the data by IMDB score
avg_gross_by_rating = avg_gross_by_rating.sort_values('imdb_score')

# Plot the data as a line graph
ax.plot(avg_gross_by_rating['imdb_score'], avg_gross_by_rating['gross'] / 1e6, '--')
ax.set_xlabel('IMDB Rating')
ax.set_ylabel('Average Gross Revenue (in millions of dollars)')
ax.set_title('Gross Revenue vs. Ratings')
plt.show()
```

Movies with higher IMDB ratings tend to have higher average gross revenues, but there is a lot of variability in the data, with some movies having high gross revenues despite lower IMDB ratings, and vice versa. The graph also shows a dip in average gross revenue for movies with an IMDB rating of around 6.5, which could indicate a sweet spot where movies are well-liked by critics but not as popular with audiences.

```python
# Group the movies by genre and calculate the total gross revenue and budget for each group
total_revenue_by_genre = df.groupby('content_rating')['gross'].mean()
total_budget_by_genre = df.groupby('content_rating')['budget'].mean()

# Create a horizontal stacked bar graph of total revenue and budget by genre
fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(total_revenue_by_genre.index, total_revenue_by_genre.values / 1e6, label='Total Gross Revenue')
ax.barh(total_budget_by_genre.index, total_budget_by_genre.values / 1e6, label='Total Budget', left=total_revenue_by_genre.values / 1e6)
plt.xlabel('Amount (in millions of dollars)')
plt.ylabel('Genre')
plt.title('Total Revenue and Budget by Genre')
plt.legend()
plt.gca().xaxis.set_major_formatter('{:.1f}M'.format)
plt.show()
```

We can see that the highest grossing content rating is PG-13, followed by PG and then R. However, PG-13 also has the highest budget, indicating that it may be more expensive to produce movies with that rating. The other content ratings have significantly lower gross revenue and budget in comparison. Overall, this suggests that producing movies with a PG-13 rating may be more financially lucrative, but also more costly.

```python
# Group the movies by year and content rating, and count the number of movies in each group
genre_count_by_year = df.groupby(['title_year', 'content_rating'])['movie_title'].count().reset_index(name='count')

# Pivot the data so that years are the rows, content ratings are the columns, and the count is the value
pivot_table = genre_count_by_year.pivot(index='title_year', columns='content_rating', values='count')
pivot_table.plot(kind='line', figsize=(10, 6))
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.title('Popularity of Genres Over Time')
plt.show()
```

The graph suggests that the number of movies released each year has increased over time for most content ratings. Additionally, it shows that the most popular content ratings for movies have changed over time. In the past, G and PG rated movies were more popular, but in recent years, PG-13 and R rated movies have become more popular. 

```python
# Group the movies by year and calculate the total gross revenue and budget for each year
total_revenue_by_year = df.groupby('title_year')['gross'].sum()
total_budget_by_year = df.groupby('title_year')['budget'].sum()

# Calculate the mean revenue and budget for each decade
mean_revenue_by_decade = total_revenue_by_year.groupby((total_revenue_by_year.index // 10) * 10).mean()
mean_budget_by_decade = total_budget_by_year.groupby((total_budget_by_year.index // 10) * 10).mean()

# Create a line graph of mean revenue and budget by decade
plt.plot(mean_budget_by_decade.index, mean_budget_by_decade / 1e9, label='Mean Budget')
plt.plot(mean_revenue_by_decade.index, mean_revenue_by_decade / 1e9, label='Mean Gross Revenue')
plt.xlabel('Year')
plt.ylabel('Amount (in billions of dollars)')
plt.title('Mean Budget and Gross Revenue by Decade')
plt.xticks(range(1920, 2021, 20))
plt.yticks(range(0, 15, 3))
plt.legend()
plt.gca().yaxis.set_major_formatter('${:.1f}B'.format)
plt.show()
```

The mean budget and gross revenue have increased over the decades, with a significant rise in the 1980s and 1990s,which can be considered as a huge change for a movie industry and then a relatively steady increase in the 2000s and 2010s. The mean budget has always been lower than the mean gross revenue, with the difference between the two increasing over the years. This suggests that while the budgets for movies have increased, they have been successful in generating more revenue over time. The trend also indicates that the movie industry has been growing steadily over the decades, with more and more money being invested in making movies, and with greater returns on that investment.

```python
# Filter the data for movies where gross revenue is at least twice the budget
df_twice_budget = df[df['gross'] >= 2 * df['budget']]

# Group the filtered data by year and calculate the count of movies in each year
twice_budget_by_year = df_twice_budget.groupby('title_year')['movie_title'].count()

# Create a line plot of the trend over the years
plt.plot(twice_budget_by_year.index, twice_budget_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Movies with Gross Revenue at least 2x Budget')
plt.title('Trend of Movies with Gross Revenue at least 2x Budget over the Years')
plt.show()
```

There is a rule of thumb is that a film would have to make at least 2x its production budget in gross revenue to become a hit.The line plot shows that the number of movies with gross revenue at least 2x the budget has been increasing over the years. There was a dip in the early 2000s, but since then, the trend has been upwards. This could be due to a variety of factors such as the popularity of certain genres or the emergence of new technologies that allow movies to be produced more efficiently. Overall, it seems that producing movies with a relatively low budget that go on to have high gross revenues is becoming more common.

```python
# Create a line plot of aspect ratio vs year
plt.plot(df.groupby('title_year')['aspect_ratio'].mean().index, df.groupby('title_year')['aspect_ratio'].mean(), '-o')

# Set the x-axis label and y-axis label
plt.xlabel('Year')
plt.ylabel('Aspect Ratio')

# Set the title of the plot
plt.title('Aspect Ratio vs Year')

# Show the plot
plt.show()
```

The aspect ratio vs year graph shows that the aspect ratio of movies has changed over time. From the early 1900s until the 1950s, the majority of movies were shot in the 4:3 aspect ratio, which is the same as old televisions. Then, widescreen aspect ratios became more common, with the 16:9 aspect ratio being the most popular in recent years.

This change in aspect ratio can be attributed to the advancements in technology and the shift in audience preferences. As TVs and movie screens became wider, filmmakers started using wider aspect ratios to enhance the cinematic experience. Additionally, audiences may have preferred the immersive experience of widescreen movies.

Overall, this graph highlights the evolution of the movie industry and how technology and audience preferences have influenced the creative decisions of filmmakers.
