# Win : loss ratio (total wins / total losses)


The win-loss ratio is calculated for each boxer, including draws by adding 0.5 to the number of draws and dividing the sum of wins, losses and draws by the total number of matches fought

```python
# Calculate the total number of fights for each stance
stance_counts = res['B_Stance'].append(res['R_Stance']).value_counts()

# Calculate the total number of wins for each stance
blue_wins =res[res['Winner'] == 'Blue']['B_Stance'].value_counts()
red_wins = res[res['Winner'] == 'Red']['R_Stance'].value_counts()
draws = res[res['Winner'] == 'Draw'][['B_Stance', 'R_Stance']].stack().value_counts()

# Combine the win counts into a single DataFrame
win_counts = pd.DataFrame({'Blue Wins': blue_wins, 'Red Wins': red_wins, 'Draws': draws})
win_counts = win_counts.fillna(0)
win_counts['Total Wins'] = win_counts['Blue Wins'] + win_counts['Red Wins'] + win_counts['Draws']

# Calculate the win ratio for each stance
win_ratios = win_counts['Total Wins'] / stance_counts

# Print the win ratios
print(win_ratios)
```

# Anova test

This code first creates a pivot table that counts the number of wins for each winning stance, and another pivot table that counts the number of fights for each stance. The two pivot tables are then merged, and the win ratio is calculated for each stance. An ANOVA model is then fit to the win ratios using the ols() function from statsmodels.formula.api, and the ANOVA test is performed using the sm.stats.anova_lm() function from statsmodels.api. The ANOVA table is printed to the console using the print() function.

The ANOVA table shows the F-statistic, the p-value, and the degrees of freedom for the model, as well as for each predictor variable (i.e., each stance). The p-value for each variable indicates whether the mean win ratio for that stance is significantly different from the mean win ratio across all stances. A p-value less than 0.05 indicates that the mean win ratio for that stance is significantly different from the mean win ratio across all stances.

```python
import sys
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols


# Create a new column for winner stance
res['winner_stance'] = res.apply(lambda x: x['B_Stance'] if x['Winner'] == 'Blue' else x['R_Stance'] if x['Winner'] == 'Red' else 'Draw', axis=1)

# Create a pivot table to count the number of wins for each stance
wins_by_stance = pd.pivot_table(res, index='winner_stance', values='Winner', aggfunc='count')

# Create a pivot table to count the number of fights for each stance
fights_by_stance = pd.pivot_table(res, index=['B_Stance', 'R_Stance'], values='Winner', aggfunc='count').reset_index()
fights_by_stance['winner_stance'] = fights_by_stance.apply(lambda x: x['B_Stance'] if x['B_Stance'] == x['R_Stance'] else 'Orthodox', axis=1)

# Merge the two pivot tables
merged = pd.merge(wins_by_stance, fights_by_stance, how='outer', on='winner_stance')

# Calculate the win ratio for each stance
merged['win_ratio'] = merged['Winner_x'] / merged['Winner_y']

# Fit an ANOVA model to the win ratios
model = ols('win_ratio ~ C(B_Stance) + C(R_Stance)', data=merged).fit()

# Perform the ANOVA test
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```

This output is from an ANOVA (Analysis of Variance) table, which is used to test for the significance of differences between groups. In this case, the ANOVA table is testing whether there is a significant difference in win ratios between different stances (B_Stance and R_Stance).

The ANOVA table has four columns:

sum_sq: the sum of squares, which is a measure of the variability in the data. It is calculated by summing the squared differences between each data point and the mean.

df: the degrees of freedom, which is the number of independent observations in the data. In this case, there are 2 degrees of freedom for each stance, since there are three stances in total.

F: the F-statistic, which is a ratio of the variability between groups to the variability within groups. It is calculated by dividing the mean square between groups by the mean square within groups.

PR(>F): the p-value, which is the probability of obtaining a result as extreme as the observed result, assuming that there is no difference between groups. A low p-value indicates that the observed result is unlikely to have occurred by chance, and therefore there is evidence of a significant difference between groups.


In this output, the p-values for both B_Stance and R_Stance are above 0.05, which means there is not enough evidence to reject the null hypothesis that there is no significant difference in win ratios between different stances. Therefore, we can conclude that there is no significant difference in win ratios between different stances.

# Anova Test

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols


# Create a new column for winner reach advantage
res['reach_advantage_winner'] = np.where(res['Winner'] == 'Blue', res['B_Reach_cms'] - res['R_Reach_cms'], res['R_Reach_cms'] - res['B_Reach_cms'])
res['reach_advantage_winner'] = np.where(res['reach_advantage_winner'] == 0, np.nan, res['reach_advantage_winner'])
res.dropna(subset=['reach_advantage_winner'], inplace=True)
res['reach_advantage_percent'] = np.where(res['reach_advantage_winner'] > 0, 1, 0)

# Log transform the reach advantage data
res['log_reach_advantage'] = np.log(res['reach_advantage_winner'])

# Fit an ANOVA model to the log transformed data
model = ols('log_reach_advantage ~ C(reach_advantage_percent)', data=res).fit()

# Perform the ANOVA test
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols



# Create a new column for reach advantage winner
res['reach_advantage_winner'] = res.apply(lambda x: 'Blue' if x['B_Reach_cms'] > x['R_Reach_cms'] and x['Winner'] == 'Blue' 
                                          or x['B_Reach_cms'] < x['R_Reach_cms'] and x['Winner'] == 'Red' 
                                          else 'Red' if x['R_Reach_cms'] > x['B_Reach_cms'] and x['Winner'] == 'Red' 
                                          or x['R_Reach_cms'] < x['B_Reach_cms'] and x['Winner'] == 'Blue' 
                                          else 'Draw', axis=1)

# Drop rows where the reach is the same
res = res[res['B_Reach_cms'] != res['R_Reach_cms']]

# Calculate the percentage of fights won by fighters with a reach advantage
reach_advantage_wins = res.groupby('reach_advantage_winner').size()
reach_advantage_total = res.groupby('reach_advantage_winner')['Winner'].count()
reach_advantage_percent = reach_advantage_wins / reach_advantage_total * 100

# Print the percentage of fights won by fighters with a reach advantage
print(reach_advantage_percent)

# Fit an ANOVA model to the percentage of fights won by fighters with a reach advantage
model = ols('reach_advantage_percent ~ C(reach_advantage_winner)', data=res).fit()

# Perform the ANOVA test
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print(anova_table)
```

#### The first part of the output shows that when one fighter had a longer reach than the other, they won 100% of the time or there was a draw 100% of the time. This suggests that having a longer reach may be an important factor in determining the outcome of a boxing match.

#### The second part of the output is an ANOVA table, which tests whether the difference in winning percentages between the two groups (longer reach vs shorter reach) is statistically significant. The table shows that the F-statistic is very large and the p-value is very small (close to 0), which indicates that there is a significant difference between the two groups in terms of winning percentages. Therefore, we can conclude that having a longer reach is likely to give a boxer an advantage in winning a match.

The output of `print(reach_advantage_percent)` shows the percentage of fights won by fighters with a reach advantage. In this case, all fights in which there was a reach advantage were won by either the Blue corner or ended in a Draw, with 100% of fights won by Blue in which there was a reach advantage.

The output of `print(anova_table)` shows the ANOVA table, which is a summary of the results of the ANOVA test that was performed on the data. The table has two rows: one for the factor being tested (`C(reach_advantage_winner)`) and one for the residual (unexplained) variance. 

The sum of squares (SS) column indicates the variability of the data that can be attributed to each factor, with higher values indicating more variability explained by the factor. In this case, the factor has a very high sum of squares value of 298.514938, indicating that there is a strong association between reach advantage and the outcome of the fight. The residual SS value of 11.761589 indicates the unexplained variability in the data after accounting for the reach advantage factor.

The degrees of freedom (df) column indicates the number of parameters estimated in the model for each factor and the residual, which is used to calculate the mean square (MS) column. The F-statistic column shows the F-value, which is the ratio of the MS for each factor to the MS for the residual. In this case, the F-value is very high for the reach advantage factor, indicating a significant effect. The p-value (PR(>F)) column indicates the probability of obtaining an F-value as extreme or more extreme than the observed value, assuming that the null hypothesis (that there is no difference between the means of the groups being compared) is true. In this case, the p-value is very small (close to zero), which indicates strong evidence against the null hypothesis and suggests that there is a significant difference between the groups.
