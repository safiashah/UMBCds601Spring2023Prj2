## 1. Data Cleaning

#### Here null values are there in these columns
I am droping them for the following reasons:

Name: What's the point in analyzing game sales if we don't have name for it so I removing the rows which do not have Name.

Year_of_Release: Though, I have game name it is difficult to search for the year_of_release because year_of_release varies for each paltform. For ex. GTA V released in year 2013 in PS3 and XBOX 360, but in PS5 and Xbox seriesX|S in year 2022.

Genre: Null for the games who has no names so removing them too.

Publisher: It is very hard to know the publisher for each game mannually.


```python
df=df.dropna(subset=['Name','Year_of_Release','Genre','Publisher'])
```

```python
df.info()
```

#### Column 'Critic_Score', 'Critic_Count', 'User_score', 'User_Count', 'Developer', and 'Rating' has too many null values and it has no contribution to the my analysis.

```python
df=df.drop(columns=['Critic_Score','Critic_Count','User_Score','User_Count','Developer','Rating'])
```

```python
df.isna().sum()
```

```python
# Here I am converting the "Year_of_Release" column into int type.

df['Year_of_Release'] = df['Year_of_Release'].astype('int64')
```

```python
# Here creating one more column called "rank" for ranking the games based on the Global sales.

df = df.sort_values('Global_Sales', ascending=False)
df['rank'] = range(1, len(df) + 1)
```

```python
df=df.reset_index()
df=df.loc[:,['rank',
             'Name',
             'Platform',
             'Year_of_Release',
             'Genre',
             'Publisher',
             'NA_Sales',
             'EU_Sales',
             'JP_Sales',
             'Other_Sales',
             'Global_Sales']]
```

## 2. EDA - Hypothesis.

### 2.1. Are shooter games published more frequently?

```python
# Plotting a graph for "Which game genre is developed more?"

genre_counts = df['Genre'].value_counts() # Counting game genres
fig = plt.bar(genre_counts.index,genre_counts.values)
plt.xticks(rotation='vertical')
plt.xlabel('Genres')
plt.ylabel('total count of each genre')
plt.title('Game genre that are developed more frequently')
plt.show()
```

```python
genre_counts
```

#### No, Action type games are developed more frequently, almost 3307 action games from year 1980 to 2020.

### 2.2. Is Super mario bros. a popular game in year 1985?

```python
# Here I am creating dictionary with year as a key and index of a game with least rank in that particular year

ranks = {}
for year in df['Year_of_Release'].unique():
    ranks[year] = df[df['Year_of_Release'] == year]['rank'].min()-1
ranks = sorted(ranks.items())
```

```python
# Hear I am creating dictionary with year as a key and game name with with least rank in that particular year

ranksV2=dict()
for i in ranks:
    ranksV2[i[0]]=df['Name'][i[1]]
ranksV2
```

   #### True, Super mario Bros. was a hit in the year it was released (i.e., 1985).

### 2.3. Activision has the highest turnover than any other publisher.

```python
# Here I am grouping each publisher with sum of their global sales and plotting them

pub_inc = df.groupby('Publisher')['Global_Sales'].sum().reset_index()

# Rearranging the dataset in descending order based on the global sales
pub_inc = pub_inc.sort_values('Global_Sales', ascending=False)
fig = plt.bar(pub_inc['Publisher'][:10],pub_inc['Global_Sales'][:10])
plt.xticks(rotation='vertical')
plt.xlabel('Publishers')
plt.ylabel('Gross income')
plt.title('Publishers VS Gross Income')
```

```python
pub_inc
```

#### False, Nitendo has a turnover of nearly 1786.72 million more than any other publishers.

### 2.4. Games are sold more in North America than Europe and Japan.

```python
# Plotting a graph for "Which continent has highest number of gamers?"

sales=dict() # Sales dictionary is for the sum of sales of all games in each continent
continents=['NA_Sales','EU_Sales','JP_Sales','Other_Sales']
for i in continents:
    sales[i]=df[i].sum()
sales
```

```python
fig = plt.bar(sales.keys(),sales.values())
plt.xticks(rotation='vertical')
plt.xlabel('continents')
plt.ylabel('Total sales in Millions')
plt.title('Sales for each continents')
plt.show()
```

#### True, just in North America, the various publishers sell games for over 4335.94 million dollars.

### 2.5. PS4 is the best gaming platform based on the revenue generated.

```python
# Plotting a graph for "which gaming platform has highest turnover"

# Here I am grouping each paltform with sum of their global sales

plat_inc = df.groupby('Platform')['Global_Sales'].sum().reset_index()

# Rearranging the dataset in descending order based on the global sales
plat_inc = plat_inc.sort_values('Global_Sales', ascending=False)
fig = plt.bar(plat_inc['Platform'],plat_inc['Global_Sales'])
plt.xticks(rotation='vertical')
plt.xlabel('Platforms')
plt.ylabel('Gross income')
plt.title('Platforms with highest revenue')
```

#### False, PS2 is best gaming platform based on the revenue generated by all games that are released in PS2 platforms.

### 2.6. Activision game publisher creates more shooter games than other genres. 

```python
#Plotting a graph for "Which publisher gets more turnover based on the genre they published?"

# Here I am grouping each publisher and genre of each publisher with sum of their global sales

grouped = df.groupby(['Publisher','Genre'])['Global_Sales'].sum().reset_index()
pivot = grouped.pivot(index='Publisher', columns='Genre', values='Global_Sales') # pivoting the dataframe
sums = pivot.sum(axis=1)

# rearranging the table in such a way that global sales sum is in decending order
sorted_df = pivot.loc[sums.sort_values(ascending=False).index]
sorted_df[0:10].plot(kind='bar', stacked=True, figsize=(15, 8))
plt.title("Best Publisher in Each Genre")
plt.xlabel("Publisher")
plt.ylabel("Global Sales (in millions)")
plt.show()
```

#### True, Activision games'  are more of a shooter genre. For example, call of duty, destiny 2 etc., are few of popular shooter games.

### 2.7. Ubisoft has published highest no. of games than anyother games

```python
#Plotting a graph for "Top 15 publishers who has published more games"

publishers_counts = df['Publisher'].value_counts() # Counting the games that each game publisher has published so far
fig = plt.bar(publishers_counts.index[:15],publishers_counts.values[:15])
plt.xticks(rotation='vertical')
plt.xlabel('publishers')
plt.ylabel('Total games published')
plt.title('Top 15 trending game publishers')
plt.show()
```

```python
publishers_counts
```

#### False, According to the graph above, Electronic Arts has created nearly 1344 games, whereas Ubisoft has created 930.

### 2.8. Is 2013 the golden year of the game industry?

```python
#Plotting a graph for "Year VS Global sales of games in each year"

# Grouping the year_of_release with the sum of global sales of all games
year_inc = df.groupby('Year_of_Release')['Global_Sales'].sum().reset_index() 
plt.plot(year_inc['Year_of_Release'], year_inc['Global_Sales'])
plt.xlabel('Year_of_Release')
plt.ylabel('Global Sales (in millions)')
plt.title('Trends Over Time in the Video Game Industry')
plt.show()
```

#### No, from year 2007 to 2010 is the golden years  in the gaming industry and year 2008 is the peak time.
