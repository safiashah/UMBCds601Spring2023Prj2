#### Reading the Dataset into 2 different  pandas dataframes : df1 and df2 in order to explain the graphs detail.

```python
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
import urllib.request
```

```python
url = "https://www.epa.gov/system/files/other-files/2022-10/emissions_by_unit_and_fuel_type_c_d_aa_10_2022.zip"
zipfile_path, _ = urllib.request.urlretrieve(url)
with zipfile.ZipFile(zipfile_path, 'r') as zip_ref:
    zip_ref.extractall()
```

#### We will use Skiprows because first 5 to 6 rows are used for describing the data.

```python
excel_file = "emissions_by_unit_and_fuel_type_c_d_aa_10_2022.xlsb"

df1 = pd.read_excel(excel_file,engine='pyxlsb',skiprows=range(0,6),sheet_name='UNIT_DATA')
df2 = pd.read_excel(excel_file,engine='pyxlsb',skiprows=range(0,5),sheet_name='FUEL_DATA')
```

#### Using seaborn plotting the graph to find out the total emission of gases.

#### melt() method is the most flexible to reshape the dataframe.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(12,12)})


df_melt = df1.melt(id_vars=['Unit Type'], 
                  value_vars=['Unit CO2 emissions (non-biogenic) ', 'Unit Methane (CH4) emissions ',
       'Unit Nitrous Oxide (N2O) emissions ',
       'Unit Biogenic CO2 emissions (metric tons)'],
                  var_name='Emission Type', value_name='Emission (metric tons)')

sns.barplot(x='Unit Type', y='Emission (metric tons)', hue='Emission Type', data=df_melt)

plt.title('Emission by unit Type')
plt.xlabel('Unit Type')
plt.ylabel('Emission (metric tons)')

plt.xticks(rotation=90)

plt.legend(title='Emission Type')

plt.show()
```

```python
common_cols = list(set(df1.columns).intersection(set(df2.columns)))

print(common_cols)
```

### Hypothesis 1: Based on Fuel Type, emissions also changes

```python
grouped_df1 = df2.groupby('General Fuel Type')['Fuel Methane (CH4) emissions (mt CO2e)'].sum()
grouped_df2 = df2.groupby('General Fuel Type')['Fuel Nitrous Oxide (N2O) emissions (mt CO2e)'].sum()

# creating 2 subplots for the graphs
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))


# Plot a bar chart of Mehtane emissions by fuel type
ann1=grouped_df1.plot(kind='bar',ax=ax1)
ax1.set_title('Methane Emissions based on fuel type')
ax1.set_xlabel('Fuel Type')
ax1.set_ylabel('CH4 Emissions')

for i, v in enumerate(grouped_df1):
    ann1.annotate(str(round(v, 2)), xy=(i, v), ha='center', va='bottom')
    
# Plot a bar chart of Nitrous Oxide emissions by fuel type

ann2= grouped_df2.plot(kind='bar',ax=ax2)
ax2.set_title('Nitrous Oxide Emissions')
ax2.set_xlabel('Fuel Type')
ax2.set_ylabel('NO2 Emissions')

for i, v in enumerate(grouped_df2):
    ann2.annotate(str(round(v, 2)), xy=(i, v), ha='center', va='bottom')
    
plt.tight_layout()
plt.show()
```

```python
grouped_df = df2.groupby('General Fuel Type')[['Fuel Methane (CH4) emissions (mt CO2e)', 'Fuel Nitrous Oxide (N2O) emissions (mt CO2e)']].sum()

# Creating a bar plot for each fuel type with the emissions for methane and nitrous oxide side-by-side
ax=grouped_df.plot(kind='bar', figsize=(10,5))


# Setting the axis labels and title
plt.xlabel('Fuel Type')
plt.ylabel('Emissions (metric tons)')
plt.title('Methane and Nitrous Oxide Emissions by Fuel Type')
    
plt.show()
```

##### _Proven Point: From the Above Graph observations we can say that the based on the type of fuel used by the industry, emission of the gases also changes_

### Hypothesis 2: Relationship between Global Warming gases(carbon-di-oxide, methane and nitrous oxide are some of the global warming gases) and increasing in number of industries by the year.

```python
df1['Reporting Year'].unique()
```

```python
sns.set(style="whitegrid")
sns.set(rc={'figure.figsize':(12,12)})


df_melt = df1.melt(id_vars=['Reporting Year'], 
                  value_vars=['Unit CO2 emissions (non-biogenic) ', 'Unit Methane (CH4) emissions ',
       'Unit Nitrous Oxide (N2O) emissions ',
       'Unit Biogenic CO2 emissions (metric tons)'],
                  var_name='Emission Type', value_name='Emission (metric tons)')

sns.barplot(x='Reporting Year', y='Emission (metric tons)', hue='Emission Type', data=df_melt)

plt.title('Emission by Fuel Type')
plt.xlabel('Reporting Year')
plt.ylabel('Emission (metric tons)')

plt.xticks(rotation=90)

plt.legend(title='Emission based on Reporting Year')

plt.show()
```

```python
grouped_df=df1.groupby(df1['Reporting Year'])['Industry Type (sectors)'].nunique()
grouped_df.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Number of Industries')
plt.title('Total Number of Industries per Year')
plt.show()
```

#### We can say that increase in industries in this sector indicates that the emission of the global warming gases is nearly constant. There might be other factors which effect the global warming.

There is a decrement in carbon emission(biogenic), which might be an effect on global warming too.
