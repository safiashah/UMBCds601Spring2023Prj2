1. Hypothesis: Clients with different income types have significantly different credit amounts for their vehicle loans.

```python
import pandas as pd
from scipy.stats import ttest_ind

# load the dataset into a pandas dataframe
data=pd.read_csv('C:/Users/poorn/OneDrive/Documents/Test_Dataset.csv', low_memory=False)

# Select the two groups to compare based on 'Client_Income_Type'
group1 = data[data['Client_Income_Type'] == 'Salary']['Credit_Amount'].dropna()
group2 = data[data['Client_Income_Type'] == 'Business']['Credit_Amount'].dropna()

# Perform two-sample t-test
t, p = ttest_ind(group1, group2, equal_var=False)

# Print the results of the test
print(f't-statistic: {t:.2f}')
print(f'p-value: {p:.4f}')


```

Fail to reject null hypothesis as the p value in Nan. There is no much difference between the two values.

2. Hypothesis: Clients with different client incomes have different no.of active loans

```python
import pandas as pd
from scipy.stats import chi2_contingency

# load the dataset into a pandas dataframe
data=pd.read_csv('C:/Users/poorn/OneDrive/Documents/Test_Dataset.csv', low_memory=False)

# create a contingency table of client income and active loans
contingency_table = pd.crosstab(data['Client_Income'], data['Active_Loan'])

# perform chi-square test of independence
chi2, p, dof, expected = chi2_contingency(contingency_table)

# print the results of the test
print(f'Chi-square statistic: {chi2:.2f}')
print(f'p-value: {p:.4f}')
```

the p-value is less than our chosen significance level (e.g., 0.05), we can reject the null hypothesis and conclude that there is a significant difference between the two values.
