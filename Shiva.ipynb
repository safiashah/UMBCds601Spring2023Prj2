# 1.Does the month of purchase affect the average sale price of a car?



Null Hypothesis: There is no significant difference in the average sale price of cars purchased in different months.

Alternative Hypothesis: There is a significant difference in the average sale price of cars purchased in different months.



```python
# convert 'Purchase Date' column to datetime and create 'Month' column
df['Purchase_Date'] = pd.to_datetime(df['Purchase Date'], format='%d-%m-%Y')
df['Month'] = df['Purchase_Date'].dt.month_name()

# group by month and calculate total sales
monthly_revenue = df.groupby('Month')['Sale Price'].sum()

# define order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June','July','August','September','October','November','December']

# convert 'Month' column to categorical with specified order
monthly_revenue.index = pd.Categorical(monthly_revenue.index, categories=month_order, ordered=True)

# sort by month order
monthly_revenue = monthly_revenue.sort_index()

# plot bar chart
fig, ax = plt.subplots()
monthly_revenue.plot(kind='bar', ax=ax)
ax.set_xlabel('Month')
ax.set_ylabel('Revenue (millions)')
ax.set_title('Monthly Car Revenue')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
plt.show()
```

<h3>From the above graph we can see the total number of sales in each month</h3>

# From the above figure we can say that all sales are in equal range per every month, so we can say that, there is no significant difference in the average sale price of cars purchased in different months and we will accept the null hypothesis.

2.Is there a relationship between the buyer's age and the likelihood of purchasing a new car?



Null Hypothesis: There is no significant relationship between the age of the buyer and the likelihood of purchasing a new car.

Alternative Hypothesis: There is a significant relationship between the age of the buyer and the likelihood of purchasing a new car.



```python
# Define the age groups and bin edges
age_groups = ['Under 30', '30-39', '40-49', '50-59', '60+']
bin_edges = [0, 30, 40, 50, 60, df['Buyer Age'].max()]

# Create a new column in the dataframe with the age group for each buyer
df['Age Group'] = pd.cut(df['Buyer Age'], bins=bin_edges, labels=age_groups)

# Create the countplot using seaborn
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Age Group', hue='New Car', palette='husl')
plt.xlabel('Age Group')
plt.ylabel('Number of Buyers')
plt.title('Distribution of New Car Purchases by Age Group')
plt.legend(title='New Car', loc='upper right')

plt.show()
```

```python
# split the dataframe into two groups based on the binary variable of new car and buyer age
group1 = df[df['New Car'] == True]['Buyer Age']
group2 = df[df['New Car'] == False]['Buyer Age']

# perform two-sample t-test
t_stat, p_val = stats.ttest_ind(group1, group2)


# print the results
print('t-statistic:', t_stat)
print('p-value:', p_val)
```

3.Does the color of a car affect the likelihood of a buyer purchasing a new car?



Null Hypothesis: There is no significant difference in the proportion of new car purchases made for different car colors.

Alternative Hypothesis: There is a significant difference in the proportion of new car purchases made for different car colors.

```python
# create a contingency table of the two categorical variables color and new car
cont_table = pd.crosstab(df['Color'], df['New Car'])

# perform chi-squared test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(cont_table)

# print the results
print('Chi-square statistic:', chi2_stat)
print('p-value:', p_val)
```

from the above we can say that p value is greater than 0.05, so we will accept the null hypothesis, i.e., There is no significant difference in the proportion of new car purchases made for different car colors.

# is there any relationship between discount and salesprice

Null Hypothesis: There is no significant relationship between discount and salesprice

Alteernate Hypothesis: There iss significant relationship between discount and salesprice

```python
corr_coef, p_value = pearsonr(df['Discount'], df['Sale Price']) #finding the correlation by pearson correlation

# Print results
print("Correlation coefficient:", corr_coef)
print("P-value:", p_value)

# Test for significance
alpha = 0.05
if p_value < alpha:
    print("There is a significant correlation between discount and sales price.")
else:
    print("There is not enough evidence to support the claim that there is a significant correlation.")
```
