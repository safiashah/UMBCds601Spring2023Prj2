# Hypothesis Testing

### Hypothesis-1

##### Steps of Hypothesis:

###### Step1: Frame the null Hypothesis (H0) 
###### Step2: Frame the Alternate Hypothesis (Ha)
###### Step3: Setup the confidence level(C.L)  ; lets say our data set has confidence level=95%
######              significance level(S.L)=1-C.L; S.L = 1-0.95 = 0.05
###### Step4: collect the evidence
###### Step5: collect the test statistic and p-value
###### Step6: if p=value < significance level --> Reject H0
######             if p=value > significance level --> Fail to Reject H0


Hypothesis 1: Are machines with higher RAM capacity less likely to be infected with malware compared to machines with lower RAM capacity? 

hypothesis 1 asks if machines with higher RAM capacity are less likely to be infected with malware compared to machines with lower RAM capacity, which also requires analyzing the data to see if there is a correlation between RAM capacity and malware infections.

Answer: As the correlation between 'Census_TotalPhysicalRAM' and 'HasDetections' is 0.060878 the hypothesis is wrong as there is no correlation.

```python
# mi_df['Census_TotalPhysicalRAM']   #--> continuous
# mi_df['HasDetections']     --> categorical

# if one column is continuous and one column is categorical then to heck hypothesis we have to do anova test
```

```python
from scipy.stats import f_oneway

# the above code using for loop

physicalram=mi_df['Census_TotalPhysicalRAM'].value_counts().index

p_value=f_oneway(*([mi_df[mi_df['Census_TotalPhysicalRAM']==ele]['HasDetections'] for ele in physicalram]))[1]
    
print(p_value)
```

```python
# so above p-value=0.0000000000......0012034 which is less then S.L(significance level=0.05)
# so if p-value<S.L, then reject Ho --> reject there is no co-relation
# conclusion is there is co-relation.
```

### Hypothesis-2

Hypothesis 2: Is there a correlation between the frequency of Windows updates and the probability of malware infection?

hypothesis 2 asks if machines that are updated more frequently may have fewer vulnerabilities and be less likely to be infected with malware, which also requires analyzing the data to see if there is a correlation between 'IsProtected' and malware infections.

Answer: As the correlation between 'IsProtected' and 'HasDetections' is 0.061281 the hypothesis is wrong as there is no correlation.

```python
# mi_df['IsProtected']   #--> categorical
# mi_df['HasDetections']     --> categorical

# if both columns are categorical then to check hypothesis we have to do chi-square test
```

```python
from scipy.stats import chi2_contingency

contingency_table=pd.crosstab(mi_df['IsProtected'],mi_df['HasDetections'])

p_value=chi2_contingency(contingency_table)[1]

print(p_value)
    
```

```python
# so above p-value=0.0000000000......0056389 which is less then S.L(significance level=0.05)
# so if p-value<S.L, then reject Ho --> reject there is no co-relation
# conclusion is there is co-relation.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# count the number of occurrences for each combination of categories
counts = mi_df.groupby(['IsProtected', 'HasDetections']).size().reset_index(name='counts')

# create bar plot
fig, ax = plt.subplots()
width = 0.35
ax.bar(counts['IsProtected'], counts['counts'], width, label='IsProtected')
ax.bar(counts['HasDetections'] + width, counts['counts'], width, label='HasDetections')

# add labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Counts')
ax.set_title('Bar Plot of Correlation between Categorical Column 1 and Categorical Column 2')
ax.legend()

# display plot
plt.show()
```

# Fill Null Values

```python
# QUE --> put all the continuos and categorical in two different lists

con_cols=[]
cat_cols=[]

for col in mi_df.columns:
    if mi_df[col].dtype=='int64' or mi_df[col].dtype=='float64':
        con_cols.append(col)
    
    if mi_df[col].dtype=='object':
        cat_cols.append(col)
        
print(con_cols)

print(cat_cols)
    
```

```python
# write the above code using list comprehension

con_cols=[col for col in mi_df.columns if mi_df[col].dtype=='int64' or mi_df[col].dtype=='float64']

cat_cols=[col for col in mi_df.columns if mi_df[col].dtype=='object']

print(con_cols)

print(cat_cols)
```

```python
for col in con_cols:
    x_train[col].fillna(x_train[col].mean(),inplace=True)
    x_test[col].fillna(x_train[col].mean(),inplace=True)
    
# filling continuous variables using mean
```

```python
for col in cat_cols:
    x_train[col].fillna(x_train[col].mode()[0],inplace=True)
    x_test[col].fillna(x_train[col].mode()[0],inplace=True)
    
# filling categorical variables using mode
```
