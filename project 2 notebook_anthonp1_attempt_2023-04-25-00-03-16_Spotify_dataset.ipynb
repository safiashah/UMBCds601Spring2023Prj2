```python
import pandas as pd
spotify = pd.read_csv("Spotify_dataset.csv")
spotify
```

```python
#Dropping the two columns like track id and valance since I found they have no use while determining the popularity of the songs that were played on Spotify app
```

```python
spotify = spotify.drop(['track_id','valence'],axis = 1)
spotify
```

```python
#Finding the null values in the data set
```

```python
print(spotify.isnull().sum())
```

```python
print(spotify.info())
```

```python
#Replacing the columns of strings which has null values with the mode values
```

```python
string_columns = spotify.select_dtypes(include=['object']).columns.tolist()
spotify[string_columns] = spotify[string_columns].fillna(spotify[string_columns].mode().iloc[0])
```

```python
print(spotify.isnull().sum())
```

Does the tempo of the song has an imapct on the popularity ?

```python
#Iam using the corr to identify the value of correlation between tempo of the song and popularity
corr = spotify['tempo'].corr(spotify['popularity'])
print(f"Correlation = {corr:.2f}")
```

```python
#Performing Hypothesis testing on the desired correlated columns
```

```python
#Import pandas and scipy.stats
```

```python
import pandas as pd
import scipy.stats as stats



# Calculate the correlation coefficient between tempo and popularity value - p 
corr_coef_tempo, p_value_popularity = stats.pearsonr(spotify['tempo'], spotify['popularity'])


#Calculate the correlation coefficient between the time duration of the song and popularity value - p
corr_coef1_duration_length,p_value1_popularity = stats.pearsonr(spotify['duration_ms'],spotify['popularity'])


#Calculate the correlation coefficient between the mode of the song and the popularity value - p
corr_coeff2_mode,p_value2_popularity = stats.pearsonr(spotify['mode'],spotify['popularity'])


print('P-value:', p_value_popularity)
print('P1-value',p_value1_popularity)
print('P2-value',p_value2_popularity)

# Perform a hypothesis test while considering the value of aplha as 0.05
alpha = 0.05
if p_value_popularity < alpha:
    print('Reject null hypothesis - There is a relationship between tempo and popularity')
else:
    print('Fail to reject null hypothesis - There is no relationship between tempo and popularity')
if p_value1_popularity < alpha:
    print('Reject null hypothesis - There is a relationship between Duration time and popularity')
else:
    print('Fail to reject null hypothesis -There is no relationship between duration time and popularity')
if p_value2_popularity < alpha:
    print('Reject null hypothesis - There is a relationship between mode and popularity')
else : 
    print('Fail to reject null hypothesis -There is no relationship between mode and popularity')
```

Does the song popularity depends on the mode?

```python
import pandas as pd

songs = spotify[(spotify['popularity'] >= 80)]

def count_zeros_and_ones(songs, mode):
    """
    Counts the number of 0s and 1s in a column of a pandas DataFrame.
    Returns a dictionary with keys 'zeros' and 'ones' and their respective counts.
    """
    # extract the column as a pandas Series
    col = songs[mode]
    # count the number of 0s and 1s in the column using the value_counts method
    counts = col.value_counts()
    # create a dictionary with the counts
    result = {'zeros': counts.get(0, 0), 'ones': counts.get(1, 0)}
    return result




# count the number of 0s and 1s in column 'col1'
counts = count_zeros_and_ones(songs, 'mode')
print(counts)  




    
```
